```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
embedding
```

```python
text ="I am iron man"

output=embedding.embed_query("text")

print(f"Text: {text}")
print(f"Embedding length: {len(output)}")
print(embedding)
```

```python
sentences=["A cat rests on top of the building.",
           "The cat is playing in the  buildinggarden.",
           "Children are running in the park.",
           "Kill the boy johnsnow,kill the boy"]
outpt=embedding.embed_documents("sentences")
print(outpt)
```

```python
import numpy as np
def cosine_similarity(vect1,vect2):
    dot_prdt=np.dot(vect1,vect2)
    nlz1=np.linalg.norm(vect1)
    nlz2=np.linalg.norm(vect2)
    return dot_prdt/(nlz1*nlz2)
```

```python
vectors = embedding.embed_documents(sentences)
print(vectors)
#Compute cosine similarity
for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        similarity = cosine_similarity(vectors[i], vectors[j])

        print(f"'{sentences[i]}' vs '{sentences[j]}'")
        print(f"Similarity: {similarity:.3f}\n")
```

```python
documents = [
  "Embeddings convert text into numbers so machines can understand meaning.",
  "Walking 10,000 steps daily helps improve heart health and burn calories.",
  "Christopher Nolan is known for making complex, mind-bending films.",
  "FastAPI is a modern Python framework used to build high-performance APIs.",
  "Saving money regularly helps build long-term financial stability."
]
```

```python
query="Which sentence explains turning text into numbers so that machines can understand meaning?"
```

```python
def sematic_search(query,documents,embedding,top_k=3):
    query_embedding=embedding.embed_query(query)
    document_embedding=embedding.embed_documents(documents)

    similarities=[]

    for i,doc_emd in enumerate(document_embedding):
        similarity=cosine_similarity(query_embedding,doc_emd)
        similarities.append((similarity,documents[i]))

    similarities.sort(reverse=True)
    return similarities[:top_k]
```

```python
result=sematic_search(query,documents,embedding)
result
```
