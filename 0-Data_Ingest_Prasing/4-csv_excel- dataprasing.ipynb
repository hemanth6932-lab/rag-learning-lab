# custom csvloader using the fucntion

```python
from langchain_core.documents import Document
from typing import List
import pandas as pd
```

```python
def Custom_cleaner(filepath:str)-> List[Document]:
     """
    Process CSV with intelligent document creation:
    - One Document per row
    - Human-readable content
    - Rich metadata for RAG
    """
     df=pd.read_csv("data/structured_files/products.csv")

     documents=[]

     for indx,row in df.iterrows():
          content = f"""Product Information:
          Name: {row['Product']}
          Category: {row['Category']}
          Price: ${row['Price']}
          Stock: {row['Stock']} units
          Description: {row['Description']}
          """
          #Meta data implementing
          doc = Document(
            page_content=content,
            metadata={
                "source": filepath,
                "row_index": indx,
                "product_name": row["Product"],
                "category": row["Category"],
                "price": row["Price"],
                "data_type": "product_info"
            }
        )
          
          documents.append(doc)
     return documents


```

```python
csv_docs = Custom_cleaner("data/structured_files/products.csv")

print(f"Loaded {len(csv_docs)} documents")
print("\nFirst document:")
print(csv_docs[0].page_content)
print(csv_docs[0].metadata)
```
